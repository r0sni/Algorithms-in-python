# 73. Set Matrix Zeroes

```python
matrix = [[1,1,1],[1,0,1],[1,1,1]]
```

```python
#Accepted but uses O(mn)
row,col = set(), set()
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        if matrix[i][j] == 0:
            row.add(i)
            col.add(j)
            
for i in row:
    for j in range(len(matrix[0])):
        matrix[i][j] = 0
        
for j in col:
    for i in range(len(matrix)):
        matrix[i][j] = 0
matrix
```

# 54. Spiral Matrix

```python
 matrix = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
left, right = 0, len(matrix[0])
top, bottom = 0, len(matrix)
out = []

while (left<right) and (top<bottom):
    #Get the top row
    for i in range(left,right,1):
        out.append(matrix[top][i])
    top += 1
    
    #Get the right column
    for i in range(top,bottom,1):
        out.append(matrix[i][right-1])
    right -= 1
    
    if not (left<right and top<bottom): #for column and row matrix
        break
        
    #Get the bottom row
    for i in range(right-1,left-1, -1):
        out.append(matrix[bottom-1][i])
    bottom -= 1
    
    #Get the left column
    for i in range(bottom-1,top-1,-1):
        out.append(matrix[i][left])
    left += 1
```

# 59. Spiral Matrix II

```python
n = 3
```

```python
left,right = 0,n
top,bottom = 0,n
```

```python
out = [[0 for i in range(n)] for j in range(n)]
out
```

```python
count = 0
n_element = [i for i in range(1,n**2+1)]
n_element
```

```python
while left<right and top<bottom:
    for i in range(left,right,1):
        out[top][i] = n_element[count]
        count += 1
    top += 1
    
    print(out)
    
    for i in range(top,bottom,1):
        out[i][right-1] = n_element[count]
        count += 1   
    right -= 1
    
    print(out)
    
    if not (left<right and top<bottom):
        break
        
    for i in range(right-1, left-1, -1):
        out[bottom-1][i] = n_element[count]
        count += 1
    bottom -= 1
    
    print(out)
        
    for i in range(bottom-1,top-1,-1):
        out[i][left] = n_element[count]
        count += 1
    left += 1
    
    print(out)
        
        
```
